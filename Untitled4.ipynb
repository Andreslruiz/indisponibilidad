```python
# %% [code] Importar librerías y cargar datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="talk")
%matplotlib inline

df = pd.read_excel('datos.xlsx', engine='openpyxl')

df['fecha_inicio_afectacion'] = pd.to_datetime(df['fecha_inicio_afectacion'])
df['fecha_fin_afectacion'] = pd.to_datetime(df['fecha_fin_afectacion'])

df['fecha'] = df['fecha_inicio_afectacion'].dt.date
```

```python
top10_sitios = df['sitio'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top10_sitios.values, y=top10_sitios.index, palette="viridis")
plt.title("Top 10 Sitios con Más Afectaciones")
plt.xlabel("Cantidad de Afectaciones")
plt.ylabel("Sitio")
plt.tight_layout()
plt.show()
```

```python
df['tiempo_horas'] = df['tiempo_afectacion'] / 60

top10_tiempo = df.groupby('sitio')['tiempo_horas'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top10_tiempo.values, y=top10_tiempo.index, palette="rocket")
plt.title("Top 10 Sitios con Mayor Tiempo de Afectación (Horas)")
plt.xlabel("Tiempo Total de Afectación (Horas)")
plt.ylabel("Sitio")
plt.tight_layout()
plt.show()
```

```python
regional_counts = df['regional'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=regional_counts.index, y=regional_counts.values, palette="mako")
plt.title("Cantidad de Afectaciones por Regional")
plt.xlabel("Regional")
plt.ylabel("Cantidad de Afectaciones")
plt.tight_layout()
plt.show()
```

```python
dept_counts = df['departamento'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=dept_counts.values, y=dept_counts.index, palette="Spectral")
plt.title("Top 10 Departamentos con Más Afectaciones")
plt.xlabel("Cantidad de Afectaciones")
plt.ylabel("Departamento")
plt.tight_layout()
plt.show()
```

```python
municipio_counts = df['municipio'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=municipio_counts.values, y=municipio_counts.index, palette="coolwarm")
plt.title("Top 10 Municipios con Más Afectaciones")
plt.xlabel("Cantidad de Afectaciones")
plt.ylabel("Municipio")
plt.tight_layout()
plt.show()
```

```python
proveedor_counts = df['proveedor'].value_counts().head(5)

plt.figure(figsize=(12,6))
sns.barplot(x=proveedor_counts.index, y=proveedor_counts.values, palette="Set2")
plt.title("Top 5 Proveedores con Más Afectaciones")
plt.xlabel("Proveedor")
plt.ylabel("Cantidad de Afectaciones")
plt.tight_layout()
plt.show()
```

```python
root_cause_counts = df['root_cause'].value_counts().head(5)

plt.figure(figsize=(12,6))
sns.barplot(x=root_cause_counts.index, y=root_cause_counts.values, palette="Set1")
plt.title("Top 5 Root Cause con Más Afectaciones")
plt.xlabel("Root Cause")
plt.ylabel("Cantidad de Afectaciones")
plt.tight_layout()
plt.show()
```

```python

df_time = df[(df['fecha_inicio_afectacion'] >= '2023-12-01') &
             (df['fecha_inicio_afectacion'] <= '2025-12-31')]

ts = df_time.set_index('fecha_inicio_afectacion').resample('M').size()

ts_df = ts.to_frame(name='cantidad')
ts_df = ts_df.reset_index()

ts_df['time_index'] = np.arange(len(ts_df))

coeffs = np.polyfit(ts_df['time_index'], ts_df['cantidad'], 1)
poly = np.poly1d(coeffs)

forecast_index = pd.date_range(start='2023-12-01', end='2025-12-31', freq='M')
forecast_time_index = np.arange(len(forecast_index))
forecast_values = poly(forecast_time_index)

plt.figure(figsize=(14,7))
plt.plot(ts_df['fecha_inicio_afectacion'], ts_df['cantidad'], marker='o', linestyle='-', label='Datos Observados', color='navy')
plt.plot(forecast_index, forecast_values, marker='x', linestyle='--', label='Tendencia y Predicción', color='darkorange')
plt.title("Tendencia de Afectaciones (2024 - 2025) y Predicción para 2025")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Afectaciones")
plt.legend()
plt.tight_layout()
plt.show()

mask_2025 = forecast_index.year == 2025
plt.figure(figsize=(14,7))
plt.plot(forecast_index[mask_2025], forecast_values[mask_2025], marker='o', linestyle='--', color='red')
plt.title("Predicción de Afectaciones para 2025")
plt.xlabel("Mes")
plt.ylabel("Cantidad de Afectaciones")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim
import time
from IPython.display import clear_output

municipio_counts = df.groupby(['municipio', 'departamento']).size().reset_index(name='count')

municipio_counts['lat'] = np.nan
municipio_counts['lon'] = np.nan

geolocator = Nominatim(user_agent="colombia_heatmap")

def get_location(municipio, departamento):
    try:
        location = geolocator.geocode(f"{municipio}, {departamento}, Colombia")
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        return None, None

print(len(municipio_counts))
for idx, row in municipio_counts.iterrows():
    lat, lon = get_location(row['municipio'], row['departamento'])
    municipio_counts.at[idx, 'lat'] = lat
    municipio_counts.at[idx, 'lon'] = lon
    print(idx, row['municipio'], lat, lon)
    time.sleep(0.5)
    clear_output(wait=True)

municipio_counts = municipio_counts.dropna(subset=['lat', 'lon'])
print("Municipios con coordenadas obtenidas:")
print(municipio_counts.head())

colombia_map = folium.Map(location=[4.5709, -74.2973], zoom_start=6)

heat_data = [[row['lat'], row['lon'], row['count']] for _, row in municipio_counts.iterrows()]

HeatMap(heat_data, radius=15, max_zoom=10).add_to(colombia_map)

colombia_map
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="talk")
%matplotlib inline

df['fecha_inicio_afectacion'] = pd.to_datetime(df['fecha_inicio_afectacion'])
df = df[df['fecha_inicio_afectacion'] >= '2024-01-01']

df_weekly = df.set_index('fecha_inicio_afectacion').resample('W').size().reset_index(name='afectaciones')
df_weekly.rename(columns={'fecha_inicio_afectacion': 'fecha'}, inplace=True)
df_weekly = df_weekly[df_weekly['afectaciones'] > 0].reset_index(drop=True)

plt.figure(figsize=(14,7))
sns.scatterplot(data=df_weekly, x='fecha', y='afectaciones', color='blue', label='Afectaciones Semanales')

df_weekly['time_index'] = np.arange(len(df_weekly))
coeffs = np.polyfit(df_weekly['time_index'], df_weekly['afectaciones'], 1)
trend_line = np.poly1d(coeffs)

plt.plot(df_weekly['fecha'], trend_line(df_weekly['time_index']), color='red', label='Tendencia lineal')
plt.title("Afectaciones de Servicio por Semana y Línea de Tendencia (Desde 2024)")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Afectaciones")
plt.legend()
plt.tight_layout()
plt.show()

last_index = df_weekly['time_index'].iloc[-1]
future_indices = np.arange(last_index + 1, last_index + 21)
future_dates = pd.date_range(start=df_weekly['fecha'].iloc[-1] + pd.Timedelta(weeks=1),
                             periods=20, freq='W')
future_predictions = trend_line(future_indices)

plt.figure(figsize=(14,7))
plt.plot(df_weekly['fecha'], df_weekly['afectaciones'], marker='o', linestyle='-', label='Histórico')
plt.plot(future_dates, future_predictions, marker='x', linestyle='--', color='orange', label='Predicción (20 semanas)')
plt.title("Predicción de Afectaciones en las Próximas 20 Semanas (Desde 2024)")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Afectaciones")
plt.legend()
plt.tight_layout()
plt.show()
```
